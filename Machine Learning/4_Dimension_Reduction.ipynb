# Dimension Reduction: Handwritten Digits
in this notebook we want to perfrom two forms a manifold learning on the handwritten digits dataset.
1. Locally Linear Embedding
2. Isomap

we will be using the scikit learn library for this task. we will also be using numpy and matplotlib for visualization purposes. Let's start by adding our libraries:

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.manifold import LocallyLinearEmbedding, Isomap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

### 1. Load the handwritten digits dataset
after importing the libraries, we start by importing the handwritten digits dataset from the scikit-learn library.

```python
digits = datasets.load_digits()
X, y = digits.data, digits.target
```

### 2. Preprocess the data
in this step we will scale the data to have zero mean and unit variance, which helps with manifold learning.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### 3. Apply and Visualize LLE(LinearLocalEmbedding):
in this step we will transform our high dimensional handwritten digits features into a 2 dimensional dataset so that we can visualize it. we will use scikit-learn for transformation of the data, and matplotlib and numpy for visualization.

```python
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=30, method='standard', random_state=42)
X_lle = lle.fit_transform(X)

# 6. Visualize LLE results
plt.figure(figsize=(8, 6))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y, cmap=plt.cm.get_cmap("viridis", 10), s=30)
plt.colorbar(ticks=np.arange(10))
plt.title("LLE Projection of Hand-written Digits")
plt.xlabel("LLE Component 1")
plt.ylabel("LLE Component 2")
plt.show()
```

as you can see above, this method didn't manage to seperate different numbers from each other properly. this could be due to two reasons: either the algorithm is hitting it's limitations and the hyperparameters are not properly set. or the handwritten digits were very similar two each other. for example a 7 can be very similar to 1, or a 6 or a 9 being very similar to 8. the problem is likely a combination of both.

### 4. Apply and Visualize Isomap
in this step we will be doing the exact same thing we did with the LLE algorithm. Let's visualize these handwritten digits using Isomap and visualize them. we will analyze the results afterwards.

```python
isomap = Isomap(n_neighbors=30, n_components=2)
X_isomap = isomap.fit_transform(X)

# 8. Visualize Isomap results
plt.figure(figsize=(8, 6))
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=y, cmap=plt.cm.get_cmap("viridis", 10), s=30)
plt.colorbar(ticks=np.arange(10))
plt.title("Isomap Projection of Hand-written Digits")
plt.xlabel("Isomap Component 1")
plt.ylabel("Isomap Component 2")
plt.show()
```
